```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = pd.DataFrame({
    'Sales': [120, 150, 180, 210, 250, 300, 330, 370, 400, 450],
    'Ad_Spend': [20, 25, 30, 35, 40, 50, 55, 60, 65, 70],
    'Discounts': [5, 7, 10, 12, 15, 18, 20, 22, 25, 28],
    'Customer_Visits': [300, 350, 400, 450, 500, 550, 600, 650, 700, 750],
    'Online_Reviews': [10, 15, 20, 25, 30, 35, 40, 45, 50, 55]
})


corr_matrix = data.corr()
print("Correlation Matrix:\n", corr_matrix)


plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.tight_layout()
plt.show()

ad_corr = corr_matrix.loc['Sales', 'Ad_Spend']
discount_corr = corr_matrix.loc['Sales', 'Discounts']

print("\n Correlation with Sales:")
print(f"Advertising Spend: {ad_corr:.2f}")
print(f"Discounts: {discount_corr:.2f}")

if abs(ad_corr) > abs(discount_corr):
    print("\n Advertising spend has a stronger correlation with sales.")
elif abs(ad_corr) < abs(discount_corr):
    print("\n Discounts have a stronger correlation with sales.")
else:
    print("\n Both factors have equal correlation with sales.")
```
