Data Visualization- Density Plot (KDE Plot)
Scenario:
A hospital is analyzing the blood sugar levels of 200 patients. The
medical team wants to see the overall distribution and identify if there
are one or more peaks in the data.
Question:
 Why is a density plot better than a histogram in this case?
 What does the peak of the KDE curve represent in terms of
patient health?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("diabetes.csv")
blood_sugar = df["Glucose"].dropna()

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.kdeplot(blood_sugar, fill=True, color="crimson", bw_adjust=1)
plt.axvline(x=130, color='blue', linestyle='--', label='Upper Normal Limit')
plt.axvline(x=180, color='purple', linestyle='--', label='Diabetic Threshold')
plt.title("Density Plot of Blood Sugar Levels (Glucose)", fontsize=16)
plt.xlabel("Blood Sugar Level (mg/dL)", fontsize=12)
plt.ylabel("Density", fontsize=12)
plt.legend()
plt.show()
```
