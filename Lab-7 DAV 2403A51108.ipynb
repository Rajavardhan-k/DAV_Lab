```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import os

# Set the SHAPE_RESTORE_SHX configuration option
os.environ['SHAPE_RESTORE_SHX'] = 'YES'

df_cases = pd.read_csv("covid.csv")
gdf_states = gpd.read_file("India_State_Boundary.shp")
print(gdf_states.columns) # Print columns to diagnose the error
gdf_states = gdf_states.merge(df_cases, left_on='Name', right_on='state') # Changed 'st_nm' to 'Name'

fig, ax = plt.subplots(1, 1, figsize=(12, 10))
gdf_states.plot(column='confirmed', cmap='Reds', linewidth=0.8, edgecolor='black', legend=True, ax=ax) # Changed 'Cases' to 'confirmed'

ax.set_title("COVID-19 Cases Across Indian States (2020)", fontsize=16)
ax.axis('off')
plt.show()
```
